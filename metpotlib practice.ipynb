```python
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
x1 = [1,2,3,4,5,6,7]
y1 = [0,15,44,77,33,55,100]
x2 = [1,2,3,4,5,6,7]
y2 = [0,26,23,49,54,32,100]
A1 = [1,7]
A2 = [1,100]
style.use('ggplot')
plt.plot(x1,y1,'b*--',label='sagar',linewidth=5) #legend should be define in the plt.plot(label="----")
plt.plot(x2,y2,'r*--',label='kunal',linewidth=2)
plt.plot(A1,A2,'g')
plt.title('SAGAR & KUNAL')
plt.xlabel("SUCCES")
plt.ylabel("YEARS")
plt.legend()
plt.grid(True,color='k')
plt.show()
```

```python
#bar plotting

plt.bar([1,2,3,4,5,6,7],[25,15,44,77,33,55,100],label="sagar",color='#034563')
plt.bar([1,2,3,4,5,6,7],[15,26,23,49,54,32,100],label='kunal',color='#587436')
plt.legend()
plt.title('SAGAR & KUNAL')
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.grid(color='red')
plt.show()
```
